```python
def cargo(WV):
  if WV <= 10:
    print("The price is R$ 50,00")
  elif WV >= 11 and WV <= 20:
    print("The price is R$ 80,00")
  else:
    print("The transport is not available")
  W_V = int(input("Cargo weight (in kg)" ))
  return cargo(W_V)


Weight = int(input("Cargo Weight (in kg)" ))
cargo(Weight)
```
