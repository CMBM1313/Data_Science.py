```python
mediana1 = sts.median(dataset["Temperatura"])
mediana1
```

```python
dataset.loc[(dataset["Temperatura"] < -130)|(dataset["Temperatura"] > 130), "Temperatura"] = mediana1
```

```python
mediana2 = sts.median(dataset["Umidade"])
mediana2
```

```python
dataset.loc[(dataset["Umidade"] < 0)|(dataset["Umidade"] > 100),"Umidade"] = mediana2
```

```python
dataset["Umidade"].fillna(mediana2, inplace = True)
```

```python
dataset["Temperatura"].fillna(mediana1, inplace = True)
```
